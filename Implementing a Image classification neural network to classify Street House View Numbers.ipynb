```python
import h5py as h
d=h.File('Autonomous_Vehicles_SVHN_single_grey1.h5','r')
d.keys()
```

```python
import numpy as n
x=n.concatenate((d['X_train'][:],d['X_val'][:]),axis=0)
y=n.concatenate((d['y_train'][:],d['y_val'][:]),axis=0)
xt=d['X_test'][:]
yt=d['y_test'][:]
x.shape
```

```python
import matplotlib.pyplot as p
p.imshow(x[0][:][:])
```

```python
y[0]
```

```python
import tensorflow.keras as k
y=k.utils.to_categorical(y,num_classes=10)
yt=k.utils.to_categorical(yt,num_classes=10)
```

```python
m=k.Sequential([k.layers.Flatten(input_shape=(1024,)),k.layers.Dense(512,activation='relu',kernel_initializer='he_normal'),k.layers.BatchNormalization(),k.layers.Dense(256,activation='relu',kernel_initializer='he_normal'),k.layers.BatchNormalization(),k.layers.Dense(128,activation='relu',kernel_initializer='he_normal'),k.layers.BatchNormalization() ,k.layers.Dense(64,activation='relu',kernel_initializer='he_normal'),k.layers.BatchNormalization() ,k.layers.Dense(10,activation='softmax')])
```

```python
m.compile(optimizer='adam',loss='categorical_crossentropy',metrics=['accuracy'])
h=m.fit(x,y,validation_data=(xt,yt),batch_size=200,epochs=510)
```

```python
p.plot(h.history['loss'])
p.plot(h.history['val_loss'])
p.xlabel('epochs')
p.ylabel('loss')
p.legend(['train','test'],loc='upper left')
p.show()
p.plot(h.history['accuracy'])
p.plot(h.history['val_accuracy'])
p.xlabel('epochs')
p.ylabel('accuracy')
p.legend(['train','test'],loc='upper left')
p.show()
```

training & test loss decreases with epochs which indicates the learning rate chosen is correct

the gap between the 2 curves in both plots is less which indicates the model is good fit
